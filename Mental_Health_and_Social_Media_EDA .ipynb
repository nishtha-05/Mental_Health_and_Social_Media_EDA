```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Mental_Health_and_Social_Media_Balance_Dataset.csv")
data.head()
```

```python
data = data[data['Gender'] != 'Other']
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.head()
```

```python
# calculating sleep efficency
data['Sleep Efficiency'] = data['Sleep_Quality(1-10)'] / data['Daily_Screen_Time(hrs)']
```

```python
# data.drop(columns = ['Sleep Ratio'],axis =1, inplace=True)
```

```python
 # stress to sleep ratio
data['Sleep Ratio'] = data['Stress_Level(1-10)'] / data['Sleep_Quality(1-10)']
data['Rtio interpretation']  = np.where(
    data['Sleep Ratio'] > 1 ,'Stress Dominates', 'good sleep')
```

```python
Platform = data['Social_Media_Platform'].value_counts()

Labels = ['TikTok','X (Twitter)','LinkedIn','Facebook','YouTube','Instagram']

plt.figure(figsize=(8,8))
plt.pie(Platform,autopct='%1.1f%%',labels=Labels,explode=(0.08,0,0,0,0,0),wedgeprops={'edgecolor':"black"})
plt.title('Social Media Platform Distribution')
plt.show()
```

```python
# plt.figure(figsize=(8,8))
# sns.countplot(x='Happiness_Index(1-10)' ,data=data , hue='Sleep_Quality(1-10)')
# plt.title(' Impact of sleep qualit on Happiness')
# plt.show()

sns.set_theme(style="darkgrid")

plt.figure(figsize=(14, 7)) 
ax = sns.countplot(
    x='Happiness_Index(1-10)',
    data=data,
    hue='Sleep_Quality(1-10)',
    palette='viridis'
)

for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=8, padding=3)

ax.set_title('Impact of Sleep Quality on Happiness', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Happiness Index (1 = Lowest, 10 = Highest)', fontsize=12)
ax.set_ylabel('Count of Respondents', fontsize=12)


plt.legend(title='Sleep Quality Score', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout() 
plt.show()
```
